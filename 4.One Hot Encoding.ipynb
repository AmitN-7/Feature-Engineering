# One Hot Encoding with Top Categories

```python
df['brand'].value_counts()
```

```python
counts=df['brand'].value_counts()
```

```python
df['brand'].nunique()
```

```python
threshold=100 
```

```python
counts[counts<=threshold].index #showing brand counts lesser than 100
```

```python
repl=counts[counts<=threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl,'uncommon'))#replacing repl & grouping all other brands(brand counts lesser than 100) within 'uncommon'
```
