**Dropping Constant feature:**
Removing all the columns that have constant values or less variance that doesn't help in problem statemnt. Deciding the threshhold value & based on the threshold variance , the features are removed

```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
```

```python
df=pd.read_csv('Customer_Satisfaction.csv',nrows=10000)
```

```python
df.shape
```

```python
df.head()
```

```python
X=df.drop(labels=['TARGET'], axis=1)
y=df['TARGET']
```

```python
variance_threshold=VarianceThreshold(threshold=0)
variance_threshold.fit(X)
```

```python
X.shape
```

```python
variance_threshold.get_support()  #True = Data should be kept within threshold value i.e, variance=0
```

```python
sum(variance_threshold.get_support())#gives the sum of columns whose variance is below zero
```

```python
total_constant_feature= X.shape[1]-sum(variance_threshold.get_support())
total_constant_feature
```

```python
columns_to_drop= X.columns[~variance_threshold.get_support()]
```

```python
X.shape
```

```python
len(columns_to_drop)
```

```python
X.drop(columns=columns_to_drop, axis=1)
```
